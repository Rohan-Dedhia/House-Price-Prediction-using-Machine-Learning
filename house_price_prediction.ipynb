```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
print(train.shape)
print(test.shape)
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
cols_to_drop = [
    "Alley",
    "MasVnrType",
    "FireplaceQu",
    "PoolQC",
    "Fence",
    "MiscFeature"
]

train = train.drop(cols_to_drop, axis=1)
test = test.drop(cols_to_drop, axis=1)
```

```python
train['LotFrontage'].skew()
```

```python
median_LotFrontage = train['LotFrontage'].median()
train['LotFrontage'] = train['LotFrontage'].fillna(median_LotFrontage)
test["LotFrontage"] = test["LotFrontage"].fillna(median_LotFrontage)
```

```python
cols = [
    "BsmtQual",
    "BsmtCond",
    "BsmtExposure",
    "BsmtFinType1",
    "BsmtFinType2"
]

for col in cols:
    train[col] = train[col].fillna("None")
    test[col] = test[col].fillna("None")
```

```python
mode_electrical = train['Electrical'].mode()[0]
train['Electrical']= train['Electrical'].fillna(mode_electrical)
```

```python
cols = ["GarageType", "GarageFinish", "GarageQual", "GarageCond"]

for col in cols:
    train[col] = train[col].fillna("None")
    test[col] = test[col].fillna("None")
```

```python
train["GarageYrBlt"] = train["GarageYrBlt"].fillna(0)
test["GarageYrBlt"] = test["GarageYrBlt"].fillna(0)
```

```python
median_mas = train["MasVnrArea"].median()

train["MasVnrArea"] = train["MasVnrArea"].fillna(median_mas)
test["MasVnrArea"] = test["MasVnrArea"].fillna(median_mas)
```

```python
mode_mszoning = train["MSZoning"].mode()[0]

test["MSZoning"] = test["MSZoning"].fillna(mode_mszoning)
```

```python
train = train.drop("Utilities", axis=1)
test = test.drop("Utilities", axis=1)  ## since all values are same
```

```python
mode_ext1 = train["Exterior1st"].mode()[0]
mode_ext2 = train["Exterior2nd"].mode()[0]

test["Exterior1st"] = test["Exterior1st"].fillna(mode_ext1)
test["Exterior2nd"] = test["Exterior2nd"].fillna(mode_ext2)
```

```python
cols = ["BsmtFinSF1","BsmtFinSF2","BsmtUnfSF","TotalBsmtSF","BsmtFullBath", "BsmtHalfBath"]

for col in cols:
    test[col] = test[col].fillna(0)
```

```python
mode_kitchen = train["KitchenQual"].mode()[0]

test["KitchenQual"] = test["KitchenQual"].fillna(mode_kitchen)
```

```python
mode_functional = train["Functional"].mode()[0]

test["Functional"] = test["Functional"].fillna(mode_functional)
```

```python
test["GarageCars"] = test["GarageCars"].fillna(0)
test["GarageArea"] = test["GarageArea"].fillna(0)
```

```python
mode_saletype = train["SaleType"].mode()[0]

test["SaleType"] = test["SaleType"].fillna(mode_saletype)
```

```python
train = train[train["GrLivArea"] < 4000]
```

```python
y_train = train["SalePrice"]
y_train = np.log1p(y_train)
```

```python
X_train = train.drop("SalePrice", axis=1)
```

```python
X_test = test.copy()
```

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
X_train, X_test = X_train.align(X_test, join="left", axis=1, fill_value=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train,y_train)
```

```python
lin_reg_pred = lin_reg.predict(X_test)
lin_reg_pred = np.expm1(lin_reg_pred)
lin_reg_pred
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
import numpy as np
ridge = Ridge()
parameters = {'alpha': np.logspace(-4, 2, 100)}
ridgecv = GridSearchCV(ridge, parameters, scoring='neg_mean_squared_error', cv=5)
ridgecv.fit(X_train, y_train)
print(ridgecv.best_params_)
ridge_pred = ridgecv.predict(X_test)
ridge_pred=np.expm1(ridge_pred)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso(max_iter=100000)
parameters = {'alpha': np.logspace(-4, 2, 100)}
lassocv = GridSearchCV(lasso, parameters, scoring='neg_mean_squared_error', cv=5)
lassocv.fit(X_train, y_train)
print(lassocv.best_params_)
```

```python
lasso_pred = lassocv.predict(X_test)
lasso_pred = np.expm1(lasso_pred)
```

```python
final_pred = (ridge_pred + lasso_pred) / 2

submission = pd.DataFrame({
    "Id": test["Id"],
    "SalePrice": final_pred
})

submission.to_csv("submission.csv", index=False)
```
